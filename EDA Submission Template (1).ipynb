## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
print('Initial Shape;',df.shape)
df=df.drop_duplicates()
print('after removing duplicates',df.shape)
missing=df.isnull().sum()
print('Missing Value:\n',missing[missing>0])
threshold=len(df)*0.5
df=df.dropna(axis=1,thresh=threshold)
num_cols=df.select_dtypes(include=['float64','int64']).columns
df[num_cols]=df[num_cols].fillna(df[num_cols].mean())
cat_cols=df.select_dtypes(include='object').columns
for col in cat_cols:
  df[col] = df[col].fillna(df[col].mode()[0])
print('remaining missing value:\n',df.isnull().sum().sum())
df['year']=df['iyear'].astype(int)
print('Final Shape:',df.shape)
````

````python
# Convert 'nkill' and 'nwound' to numeric, coercing errors
df['nkill'] = pd.to_numeric(df['nkill'], errors='coerce')
df['nwound'] = pd.to_numeric(df['nwound'], errors='coerce')

# Fill any newly created NaNs in 'nkill' and 'nwound' with their respective means
df['nkill'] = df['nkill'].fillna(df['nkill'].mean())
df['nwound'] = df['nwound'].fillna(df['nwound'].mean())

print("Data types for 'nkill' and 'nwound' after conversion and filling:")
print(df[['nkill', 'nwound']].dtypes)
````

### What all manipulations have you done and insights you found?

After loading, the dataset was found to have no duplicate rows. Significant data manipulation was performed to handle missing values:
1.  **Columns with >50% missing values were dropped**, reducing the number of columns from 135 to 59, to focus on more complete features.
2.  **Missing numerical values** (e.g., in `nkill`, `nwound`) were **imputed with the mean** of their respective columns.
3.  **Missing categorical values** were **imputed with the mode**.
4.  The `iyear` column was explicitly converted to an integer `year` column.
5.  `nkill` and `nwound` columns were explicitly converted to numeric (`float64`), with any conversion errors coerced to NaN, and then these new NaNs were filled with the mean of their columns. This ensures these crucial impact metrics are ready for numerical analysis.

The primary insight is that the dataset has been thoroughly cleaned and prepared, making it analysis-ready with no remaining missing values, ensuring reliability for subsequent visualization and modeling.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
````

##### 1. Why did you pick the specific chart?

Understanding these correlations helps in feature selection for predictive models and understanding underlying data structures. Strong correlations might simplify models, but also indicate potential multicollinearity. Lack of expected correlations might highlight data quality issues or suggest variables are independent, guiding further analysis.

It quickly identifies which numerical variables move together (positive correlation) or in opposite directions (negative correlation), and which have little to no linear relationship.
